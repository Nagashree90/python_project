identity operators

🔹 is checks object identity, while == checks value equality.
🔹 b = a makes b a reference to the same object as a.
🔹 c = [1, 2, 3] creates a new list, so a is c is False.



```python
a = [1, 2, 3]
b = a  # Both refer to the same list object
c = [1, 2, 3]  # A new list with the same values

print(a is b)   # True (Same memory reference)
print(a is c)   # False (Different objects, even though values are the same)
print(a == c)   # True (Values are equal)

print(a is not c)  # True (Different memory locations)
```

Checking Identity of Immutable Objects
Python optimizes memory by reusing immutable objects (e.g., integers, strings)

```python
x = 5
y = [5]
print(y)
z = 5
print(y is z)  # True (Same memory location for small integers)
print( x is z)
print(x is y)

s1 = "hello"
s3 = s1
s2 = "hello"
print(s3 is s2)  # True (Same memory location due to string interning)
```

Examples of Operator Precedence
Parentheses Have the Highest Precedence
 Parentheses () force certain calculations first.

```python
print((2 + 3) * 4)  # 5 * 4 = 20
print(2 + 3 * 4)    # 2 + (3 * 4) = 2 + 12 = 14
```

Exponentiation (**) is Evaluated Right to Left

```python
print(2 ** 3 ** 2)  # 2 ** (3 ** 2) = 2 ** 9 = 512
```

Multiplication and Division Happen Before Addition and Subtraction

```python
print(10 - 2 + 3 * 2)  # 10 - 2 + (3 * 2) = 10 - 2 + 6 = 14
print(10 / 2 * 4)      # (10 / 2) * 4 = 5 * 4 = 20
```

Logical Operators (not, and, or)
not has the highest precedence, followed by and, then or.



```python
print(True or False and False)  # True or (False and False) = True or False = True
print(not True or False)        # (not True) or False = False or False = False
```
