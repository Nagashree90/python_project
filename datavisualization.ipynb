```python
#histogram in matplotlib
import matplotlib.pyplot as plt

data = [22, 25, 26, 28, 30, 30, 32, 35, 36, 38, 40, 42, 45]

plt.hist(data, bins=5, color='orange', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--')
plt.show()
```

What does bins=5 mean?
You're asking matplotlib to divide the data into 5 equal-width intervals (called bins).

Each bin represents a range of values.

It will then count how many data points fall into each bin.

Step-by-Step:
Step 1: Find the min and max of the data
Min = 22

Max = 45

Step 2: Calculate bin width
Bin width = (Max - Min) / number of bins
= (45 - 22) / 5
= 23 / 5
= 4.6



Bin Ranges (Automatically calculated):
Bin Number	Bin Range (approx)	Includes which values
Bin 1	       22.0  26.6	       22, 25, 26
Bin 2.      	26.6  31.2	       28, 30, 30
Bin 3	        31.2  35.8	       32, 35
Bin 4	        35.8  40.4	       36, 38, 40
Bin 5	       40.4  45.0	        42, 45


The histogram uses left-inclusive, right-exclusive bins by default:

A value belongs to the bin if it's  left edge and < right edge.

The last bin includes its right edge.

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (3 groups for comparison)
group1 = [22, 25, 27, 30, 32, 34, 35, 36, 37, 38]
group2 = [18, 20, 23, 24, 28, 29, 33, 34, 35, 40]
group3 = [21, 23, 26, 29, 30, 31, 34, 35, 50, 100]  # has outliers

data = [group1, group2, group3]

# Create box plot
plt.boxplot(
    data,
    notch=True,                     # Adds a notch for median confidence interval
    vert=True,                      # True = vertical box plot
    patch_artist=True,              # Fill the boxes with color
    labels=['Group A', 'Group B', 'Group C'],  # Custom group labels
    showmeans=True,                 # Show the mean as a green triangle
    meanline=False,                 # If True, shows the mean as a line
    boxprops=dict(facecolor='skyblue', color='black'),       # Box color
    whiskerprops=dict(color='black', linestyle='--'),        # Whisker line style
    capprops=dict(color='red'),                             # Cap line color
    medianprops=dict(color='purple', linewidth=2),           # Median line style
    meanprops=dict(marker='o', markerfacecolor='green', markeredgecolor='black', markersize=8),  # Mean marker
    flierprops=dict(marker='o', markerfacecolor='red', markersize=7, linestyle='none')  # Outlier (flier) style
)

# Title and labels
plt.title("Box Plot with All Parameters")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
